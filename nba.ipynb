```python
# Load data
df3 = pd.read_csv("Seasons_Stats.csv")
df3
```

```python
curry_rows = df3[df3['Player'] == 'Stephen Curry']
print(curry_rows)
```

```python
scoring_per_position = df3.groupby('Pos')['PTS'].mean()
print(scoring_per_position)
```

```python
reb_per_position = df3.groupby('Pos')['TRB'].mean()
print(reb_per_position)
```

```python
ast_per_position = df3.groupby('Pos')['AST'].mean()
print(ast_per_position)
```

```python
efficiency_per_decade = df3.groupby('Year')['PER'].mean()
latest_decade = efficiency_per_decade.sort_index().head(10)
print(latest_decade)
```

```python
efficiency_per_decade = df3.groupby('Year')['PER'].mean()
latest_decade = efficiency_per_decade.sort_index().tail(10)
print(latest_decade)
```

```python
df3['Decade'] = (df3['Year'] // 10) * 10
efficiency_per_decade = df3.groupby('Decade')['PER'].mean()
print(efficiency_per_decade)
```
